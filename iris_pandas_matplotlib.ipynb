Summary Functions and `map` / `apply`

Summary statistics (sepal length mean, petal width median, value_counts())

```python
df['sepal_length_rounded']= df['SepalLengthCm'].map(lambda x: round(x, 1))
```

```python
df.head()
```

```python
df['setosa_code']= df['Species'].map({
    'Iris-setosa': 1,
    'Iris-virginica': 2,
    'Iris-versicolor':3
})
df.head()
```
